```python
# 匯入模組
from langchain_community.document_loaders import PyPDFLoader

# 載入PDF(表格圖片)
loader = PyPDFLoader(file_path="/content/表格圖片.pdf", extract_images=True)

# 轉換成Document物件
docs = loader.load()
print(docs)
```

```python
# 免費Embedding模型 - HuggingFace
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.indexes import VectorstoreIndexCreator

embeddings_model = HuggingFaceEmbeddings(model_name="intfloat/multilingual-e5-large")

index = VectorstoreIndexCreator(embedding=embeddings_model).from_loaders([loader])
```

```python
query = "PDF中的內容，主要在講甚麼"
response = index.query(llm=chat_model, question=query)
print(response)
```
