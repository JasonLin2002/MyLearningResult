```python
import os

def traverse_directory(directory_path):
    # 遍歷資料夾中的所有檔案和子資料夾
    for root, dirs, files in os.walk(directory_path):
        print(f"目前資料夾: {root}")
        for file in files:
            print(f"檔案: {os.path.join(root, file)}")
        for dir in dirs:
            print(f"子資料夾: {os.path.join(root, dir)}")

# 使用範例
directory_path = 'data'
traverse_directory(directory_path)
```

# 遍歷所有子資料夾中的檔案，以印出 html 檔案內容

```python

import os
from bs4 import BeautifulSoup

pic_user=[]
raw_text=[]


def print_html_per_directory(directory_path):
    for root, dirs, files in os.walk(directory_path):
        print(f"目前資料夾: {root}")
        for file in files:
            if file.endswith(".html"):
                file_path = os.path.join(root, file)
                
                # 讀取 HTML 檔案內容
                with open(file_path, 'r', encoding='utf-8') as html_file:
                    content = html_file.read()
                    
                    # 使用 BeautifulSoup 解析 HTML 檔案
                    soup = BeautifulSoup(content, 'lxml')
                    
                    # 提取文本
                    text = soup.get_text()
                    
                    print(f"Text content of {file}:\n")
                    print(text)
                    print("=" * 50)
                    text =  text.strip()
                    if len(text)>1:
                        raw_text.append(text)
            if file.endswith(".png"):
                pic_user.append(str(root))
                
    print("有{}個同學有以圖片的方式繳交".format(len(pic_user)))



# 使用範例
directory_path = '1131_0913'
print_html_per_directory(directory_path)
```
